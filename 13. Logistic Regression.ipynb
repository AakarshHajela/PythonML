```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
data = pd.read_csv('data_cleaned.csv')
data.head()
```

```python
x = data.drop(['Survived'], axis = 1)
y = data['Survived']
```

```python
from sklearn.model_selection import train_test_split as tts
train_x, test_x, train_y, test_y = tts(x, y, random_state = 56)
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
```

```python
cols = train_x.columns
cols
```

```python
train_x_scaled = scaler.fit_transform(train_x)
train_x_scaled = pd.DataFrame(train_x_scaled, columns = cols)
train_x_scaled.head()
```

```python
test_x_scaled = scaler.transform(test_x)
test_x_scaled = pd.DataFrame(test_x_scaled, columns = cols)
test_x_scaled.head()
```

 # Implementing Logistic Regression

```python
from sklearn.linear_model import LogisticRegression as LogReg
from sklearn.metrics import f1_score 
```

```python
logreg = LogReg()
logreg.fit(train_x, train_y)
```

```python
train_predict = logreg.predict(train_x)
train_predict
```

```python
k = f1_score(train_predict, train_y)
print('Train f1 score ', k)
```

```python
test_predict = logreg.predict(test_x)
k = f1_score(test_predict, test_y)
print('Test f1 score ', k)
```

# Predicting Probabilities

```python
train_predict = logreg.predict_proba(train_x)
train_predict
```

Checking whether all Probabilities are between 0 and 1

```python
train_preds = train_predict[:,1]
train_preds
```

```python
for i in range(0, len(train_preds)):
    if(train_preds[i]>0.55):
        train_preds[i] = 1
    else:
        train_preds[i] = 0
```

```python
k = f1_score(train_preds, train_y)
print('Training F1 Score ', k)
```

# Confusion Matrix

```python
from sklearn.metrics import confusion_matrix
cf = confusion_matrix(test_y, test_predict)
print(cf)
```

```python
from sklearn.metrics import classification_report as rep
print(rep(test_y, test_predict))
```

# Plotting the  Coefficients

```python
plt.figure(figsize = (8,6), dpi = 120, facecolor = 'w', edgecolor = 'b')
x = range(len(train_x.columns))
c = logreg.coef_.reshape(-1)
plt.bar(x,c)
plt.xlabel = 'Variables'
plt.ylabel = 'Coefficients'
plt.title = 'Coefficient Plot'
```

```python
Coefficients = pd.DataFrame({'Variables' : train_x.columns, 'Coefficients':abs(c)})
Coefficients.head()
```

```python
sig_var = Coefficients[Coefficients.Coefficients > 0.3]
```

```python
subset = data[sig_var['Variables'].values]
subset.head()
```
