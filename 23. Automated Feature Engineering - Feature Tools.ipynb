```python
import pandas as pd
import featuretools as ft
```

```python
df = pd.read_csv('train_bm.csv')
df.head()
```

```python
features = df.drop(['Item_Outlet_Sales'], axis = 1)
y = df['Item_Outlet_Sales']
```

## Make an Entity Set and add the entity

```python
es = ft.EntitySet(id = 'Bigmart')
es
```

## Adding data to the Entity Set created

```python
es.entity_from_dataframe(entity_id = 'data_1', dataframe = features, make_index = True, index = 'index')
es
```

# Feature Engineering

```python
feature_matrix, feature_defs = ft.dfs(entityset = es, target_entity = 'data_1', max_depth = 1,
                                      trans_primitives = ['add_numeric','multiply_numeric'])
feature_matrix.head()
```
